```python
from carte_ai.data import lung_cancer
```

```python
data = lung_cancer.read_lung_cancer()

data.head()
```

All numerical values are almost perfectly uniformly distributed

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = data.select_dtypes(include=['number'])
for col in numerical_cols.columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(data=data, x=col, kde=True, bins=30, color='blue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(axis='y')
    plt.show()
```

More surprising, is that there seems to be absolute no correlation between any of the columns. Even things you'd suspect to be correlated like 'Age' and 'Survival_Months'.

My hunch is that this dataset is synthetic + made of completely independently random values, and that there are no patterns to learn from it.

```python
numerical_cols = data.select_dtypes(include=['number'])
correlation_matrix = numerical_cols.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, annot_kws={"size": 8}, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Pairwise Correlation Heatmap")
plt.show()
```
